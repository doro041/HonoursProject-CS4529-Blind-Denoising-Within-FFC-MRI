```python
import numpy as np
from scipy.io import loadmat
from skimage import io
import os

# Specify the folder containing the .mat files
folder_path = '../noise2fast/SimulationMatlab'

# List all .mat files in the folder
file_list = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.mat')]

for file in file_list:
    # Load the .mat file
    data = loadmat(file)
    
    # Extract 4D data
    if 'data_gaussian' in data:
        data_4d = data['data_gaussian']  # For Gaussian data
    elif 'data_rician' in data:
        data_4d = data['data_rician']  # For Rician data
    else:
        print(f"Unknown data type in file: {file}")
        continue
    
    # Iterate through each magnetic field (3rd dimension)
    for m in range(data_4d.shape[3]):
        # Then iterate through each time frame (4th dimension)
        for t in range(data_4d.shape[2]):
            # Extract the magnetic field data for the current slice
            slice_2d = data_4d[:, :, t,m]
            
            # Normalize the magnetic field data to be in the range [0, 255]
            normalized_slice = (slice_2d - slice_2d.min()) * 255 / (slice_2d.max() - slice_2d.min())
            
            # Convert the data to uint8
            uint8_slice = normalized_slice.astype('uint8')
            
            # Generate the file name based on magnetic field and time frame
            tiff_file_name = f"{os.path.splitext(os.path.basename(file))[0]}_magfield_{m}_frame_{t}.tif"
            tiff_file_path = os.path.join(os.path.dirname(file), tiff_file_name)
            
            # Save the slice as a TIFF file
            io.imsave(tiff_file_path, uint8_slice)
            print(f"Saved: {tiff_file_path}")
```
