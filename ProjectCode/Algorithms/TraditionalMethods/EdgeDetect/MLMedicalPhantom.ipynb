```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage import feature
from skimage.color import rgba2rgb
from skimage.color import rgb2gray

def apply_canny_edge_detection(image_path, sigma=1.0):
    """
    Apply Canny edge detection to a given image.
    
    :param image_path: Path to the image file.
    :param sigma: Standard deviation of the Gaussian filter used in Canny edge detector.
    :return: Edge-detected image.
    """
    image = imread(image_path)
    
    # Handle images with 4 channels (RGBA) by converting to RGB
    if image.ndim == 3 and image.shape[2] == 4:
        image = rgba2rgb(image)
    
    # Convert image to grayscale if it's not already
    if image.ndim == 3:
        image = rgb2gray(image)
    
    # Apply Canny edge detector
    edges = feature.canny(image, sigma=sigma)
    
    return edges



def visualize_edges(image_paths, sigma=1.0):
    """
    Visualize the edges detected in all images in a single figure.
    
    :param image_paths: List of paths to the image files.
    :param sigma: Standard deviation of the Gaussian filter used in Canny edge detector.
    """
    num_images = len(image_paths)
    nrows = int(np.ceil(np.sqrt(num_images)))
    ncols = nrows
    
    plt.figure(figsize=(15, nrows * 5))
    
    for i, image_path in enumerate(image_paths, start=1):
        edges = apply_canny_edge_detection(image_path, sigma=sigma)
        plt.subplot(nrows, ncols, i)
        plt.imshow(edges, cmap='gray')
        
        title = image_path.split('/')[-1]  # Extract the file name as title
        plt.title(title)
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()

# Example usage
image_paths = [
    '../Experiments/DenoisedImages//MedicalPhantomTrad/MedicalPhantom_N2F/200 mT_116 ms_slice_1.tif',
    '../Experiments/DenoisedImages/MedicalPhantomTrad/MedicalPhantom_patch2self/denoised_200mT.tiff',
    '../Experiments/DenoisedImages/MedicalPhantomTrad/MedicalPhantom_noise2self/200 mT_116 ms_slice_1.tif'
]

visualize_edges(image_paths, sigma=1.5)  # Adjust sigma as needed for your images

```
