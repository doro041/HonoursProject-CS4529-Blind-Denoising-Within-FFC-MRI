### Develop a script to transform a series of 2D medical phantom images into a 3D volume. Given that there are five images, each with dimensions of 50 by 100 pixels, the goal is to stack these images along a third dimension representing different evolution times, resulting in a 3D array with dimensions 50x100x5.

```python
import imageio
import numpy as np

# List of TIFF filenames
filenames = ['../ProjectCode/Algorithms/TraditionalMethods/PhantomData/200mT/200mT_116 ms_slice_1.tif', '../ProjectCode/Algorithms/TraditionalMethods/PhantomData/200mT/200mT_229 ms_slice_1.tif', '../ProjectCode/Algorithms/TraditionalMethods/PhantomData/200mT/200mT_451 ms_slice_1.tif', '../ProjectCode/Algorithms/TraditionalMethods/PhantomData/200mT/200mT_59 ms_slice_1.tif',  '../ProjectCode/Algorithms/TraditionalMethods/PhantomData/200mT/200mT_887 ms_slice_1.tif']

# Load TIFF images and store them in a list
image_data = []
for filename in filenames:
    img = imageio.imread(filename)
    image_data.append(img)

# Convert the list of images to a NumPy array
image_data_array = np.array(image_data)

# Display shape of the NumPy array (number of images, height, width)
print("Shape of image data array:", image_data_array.shape)

# Save the NumPy array to a file
np.save('image_data.npy', image_data_array)
```

```python
import matplotlib.pyplot as plt

# Load the NumPy array from the saved file
loaded_image_data = np.load('image_data.npy')

# Combine the images along the third axis (time axis)
stacked_image = np.sum(loaded_image_data, axis=0)

# Display the stacked 3D image
plt.imshow(stacked_image, cmap='gray')

plt.axis('off')

# Save the image
plt.savefig('200mTPhantom.tif', bbox_inches='tight')
plt.show()


```
